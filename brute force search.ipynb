```python
import pandas as pd
from itertools import permutations

# Loading and cleaning matrix
def load_matrix(path):
    df = pd.read_csv(path, header=None)
    matrix = df.iloc[1:, 1:].astype(int).values.tolist()
    return matrix

# Distance calculation
def calculate_distance(route, matrix):
    return sum(matrix[route[i]][route[i+1]] for i in range(len(route)-1)) + matrix[route[-1]][route[0]]

# Brute-force search
def brute_force_search(matrix):
    cities = list(range(1, len(matrix)))
    best_route = None
    min_distance = float('inf')
    for perm in permutations(cities):
        route = [0] + list(perm)
        dist = calculate_distance(route, matrix)
        if dist < min_distance:
            best_route = route
            min_distance = dist
    return best_route, min_distance

# Running/testing
if __name__ == "__main__":
    matrix = load_matrix("distances.csv")
    route, distance = brute_force_search(matrix)
    print("Brute-force Search Result:")
    print("Route:", route)
    print("Total Distance:", distance)
```
